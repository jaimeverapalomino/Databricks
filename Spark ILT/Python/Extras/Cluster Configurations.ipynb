#Presentation Mode
To enable presentation mode, open the JavaScript console and enter the following command:

`window.settings.enablePresentationMode = true`

Then from the **View** menu select **Present**

      <li>Highly repetitive queries</li>

<div style="display:table; height:600px; width:100%"><div style="display:table-cell; vertical-align:middle">

    <div>Data Scientist</div>
    <ul style="margin-left:1em">
      <li>Large overlap with Data Analyst</li>
      <li>Statistical perspective to data</li>
      <li>Highly repetitive queries (at times)</li>
      <li>Engage in data preparation (regularly)</li>
      <li>Train models w/highly iterative jobs</li>
    </ul>
  </div>
</div></div>

      <li>Strong concern for execution time</li>
      <li>Building data pipelines (dirty &rarr; clean)</li>
      <li>Productionalizing ML Models and reports</li>
      <li>Non repetitive, scheduled queries
