## Create a Runnable Notebook

[Fill out the notebook in this directory 04-Lab-Runnable.]($./04-Lab-Runnable)  It should accomplish the following:<br><br>

0. Read in 3 parameters:
  - `n_estimators` with a default value of 100
  - `learning_rate` with a default value of .1
  - `max_depth` with a default value of 1
0. Train and log a gradient boosted tree model
0. Report back the path the model was saved to

Run the notebook with the following code.

```
dbutils.notebook.run("./04-Lab-Runnable", 60, 
  {"n_estimators": "100",
   "learning_rate": ".1",
   "max_depth": "1"})
```
