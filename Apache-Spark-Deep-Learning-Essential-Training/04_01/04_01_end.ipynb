### Working with images
##### Check that we have correct libraries installed

```python
import tensorflow
import keras
import h5py
assert(tensorflow.__version__ == '1.12.0')
assert(keras.__version__ == '2.2.4')
assert(h5py.__version__ == '2.7.0')
```

##### Import image into dataframe

```python
img  = '/FileStore/tables/typewriter.jpg'

```

```python
from pyspark.ml.image import ImageSchema
df = ImageSchema.readImages(img)
df.show()
df.printSchema()
```

##### Using pre-trained models

```python
from sparkdl import DeepImagePredictor
```

```python
predictor = DeepImagePredictor(inputCol='image',outputCol='predicted_labels', modelName='ResNet50', decodePredictions=True, topK=5)
```

```python
predictor_df = predictor.transform(df)
predictor_df.select('predicted_labels').show(truncate=False)
```
